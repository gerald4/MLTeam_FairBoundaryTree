```python
from FAHT import cart_tree_FAHT
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
np.load('y_adult.npy').astype(np.intp)
```

```python
Y = np.loadtxt("datasets/compas/compas_Y.csv")
Y = Y.astype("int64")
X = pd.read_csv("datasets/compas/compas_X.csv", index_col=0)
colnames = list(X)
X = X.values
```

```python
dt = cart_tree_FAHT.DecisionTreeFAHT(min_samples = 1, impurity_type = "gini", _lambda = 2., max_leaves = 6)


X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.30, random_state = 111)

dt.fit(X_train, y_train)
```

```python
print("Predictive accuracy: ", accuracy_score(dt.predict(X_test), y_test))
```

```python
def computeDI(X, y):
    table1 = np.bincount(y[np.where(X[:,-1] == 0)[0]])
    table2 = np.bincount(y[np.where(X[:,-1] == 1)[0]])
    
    return cart_tree_FAHT.unfairness(table1, table2)
```

```python
def computeDM(X, y, y_pred):
    table1 = np.bincount(y[np.where(X[:,-1] == 0)[0]])
    table2 = np.bincount(y[np.where(X[:,-1] == 1)[0]])
    
    return cart_tree_FAHT.unfairness(table1, table2)
```
